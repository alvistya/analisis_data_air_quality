### Cleaning Data

```python
print('Stasiun Tiantan\n')

print('Mengonversi kolom year, month, day, hour menjadi datetime')
st_tiantan['datetime'] = pd.to_datetime(st_tiantan[['year', 'month', 'day', 'hour']])

print('Menghapus kolom yang sudah tergabung ke datetime')
st_tiantan.drop(columns=['year', 'month', 'day', 'hour', 'No'], inplace=True, errors='ignore')

print('Mengisi data kosong pada kolom numerik dengan interpolasi')
st_tiantan.loc[:, st_tiantan.select_dtypes(include=['number']).columns] = (
    st_tiantan.select_dtypes(include=['number']).interpolate(method='linear').round(1)
)

print('Mengisi data kosong pada kolom kategori dengan modus')
st_tiantan['wd'].fillna(st_tiantan['wd'].mode()[0], inplace=True)

print('Menyimpan dataset yang sudah bersih')
st_tiantan.to_csv('dashboard/cleaned_Tiantan.csv', index=False)
```

```python
print('Stasiun Wanliu\n')

print('Mengonversi kolom year, month, day, hour menjadi datetime')
st_wanliu['datetime'] = pd.to_datetime(st_wanliu[['year', 'month', 'day', 'hour']])

print('Menghapus kolom yang sudah tergabung ke datetime')
st_wanliu.drop(columns=['year', 'month', 'day', 'hour', 'No'], inplace=True, errors='ignore')

print('Mengisi data kosong pada kolom numerik dengan interpolasi')
st_wanliu.loc[:, st_wanliu.select_dtypes(include=['number']).columns] = (
    st_wanliu.select_dtypes(include=['number']).interpolate(method='linear').round(1)
)

print('Mengisi data kosong pada kolom kategori dengan modus')
st_wanliu['wd'].fillna(st_wanliu['wd'].mode()[0], inplace=True)

print('Menyimpan dataset yang sudah bersih')
st_wanliu.to_csv('dashboard/cleaned_Wanliu.csv', index=False)
```

```python
print('Stasiun Shunyi\n')

print('Mengonversi kolom year, month, day, hour menjadi datetime')
st_shunyi['datetime'] = pd.to_datetime(st_shunyi[['year', 'month', 'day', 'hour']])

print('Menghapus kolom yang sudah tergabung ke datetime')
st_shunyi.drop(columns=['year', 'month', 'day', 'hour', 'No'], inplace=True, errors='ignore')

print('Mengisi data kosong pada kolom numerik dengan interpolasi')
st_shunyi.loc[:, st_shunyi.select_dtypes(include=['number']).columns] = (
    st_shunyi.select_dtypes(include=['number']).interpolate(method='linear').round(1)
)

print('Mengisi data kosong pada kolom kategori dengan modus')
st_shunyi['wd'].fillna(st_shunyi['wd'].mode()[0], inplace=True)

print('Menyimpan dataset yang sudah bersih')
st_shunyi.to_csv('dashboard/cleaned_Shunyi.csv', index=False)
```

## Exploratory Data Analysis (EDA)

### Explore ...

Saya mulai dengan mengimpor dataset yang telah dibersihkan dari folder dashboard. Dataset ini berasal dari tiga lokasi berbeda: Tiantan, Shunyi, dan Wanliu. Saya juga memastikan bahwa kolom datetime diubah menjadi format tanggal agar dapat dikelompokkan berdasarkan tahun.

```python
print('Mengimpor atau mengambil data yang sudah di berishkan')
st_cleaned_tiantan = pd.read_csv('dashboard/cleaned_Tiantan.csv', parse_dates=['datetime'])
st_cleaned_shunyi = pd.read_csv('dashboard/cleaned_Shunyi.csv', parse_dates=['datetime'])
st_cleaned_wanliu = pd.read_csv('dashboard/cleaned_Wanliu.csv', parse_dates=['datetime'])
```

Saya kemudian menentukan enam jenis polutan utama yang akan digunakan untuk analisis perbandingan, yaitu PM2.5, PM10, SO2, NO2, CO, dan O3.

Saya mengelompokkan data berdasarkan tahun dan menghitung rata-rata tahunan dari masing-masing polutan untuk setiap stasiun. Dengan cara ini, saya bisa melihat bagaimana kualitas udara berubah dari tahun ke tahun.

```python
print('Memilih parameter kualitas udara utama untuk perbandingan')
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

print('Hitung rata-rata tahunan untuk setiap stasiun')#
st_tiantan_avg = st_cleaned_tiantan.groupby(st_cleaned_tiantan['datetime'].dt.year)[pollutants].mean()
st_shunyi_avg = st_cleaned_shunyi.groupby(st_cleaned_shunyi['datetime'].dt.year)[pollutants].mean()
st_wanliu_avg = st_cleaned_wanliu.groupby(st_cleaned_wanliu['datetime'].dt.year)[pollutants].mean()
```

Saya menyusun hasil perhitungan tadi ke dalam sebuah tabel baru (comparison_df) agar lebih mudah dibandingkan antar stasiun. Setiap baris mewakili satu stasiun, sedangkan kolomnya menunjukkan rata-rata polutan selama periode yang diamati.

```python
print('Buat DataFrame baru untuk perbandingan')
comparison_df = pd.DataFrame({
    'Tiantan': st_tiantan_avg.mean(),
    'Shunyi': st_shunyi_avg.mean(),
    'Wanliu': st_wanliu_avg.mean()
}).T
```

Berdasarkan proses Exploratory Data Analysis (EDA) yang telah dilakukan terhadap data kualitas udara dari tiga stasiun pengukuran (Tiantan, Shunyi, dan Wanliu), berikut adalah beberapa wawasan yang dapat diperoleh:

**Insight:**
1. Tren Rata-rata Polutan Tahunan
 - Dengan menghitung rata-rata tahunan dari setiap polutan (PM2.5, PM10, SO2, NO2, CO, O3), dapat diketahui apakah tingkat polusi mengalami peningkatan atau penurunan dari tahun ke tahun.
 - Jika ada tren penurunan, mungkin ada kebijakan lingkungan yang berdampak positif, seperti pengurangan emisi kendaraan atau peningkatan ruang hijau.
 - Jika ada tren peningkatan, hal ini bisa mengindikasikan peningkatan aktivitas industri atau urbanisasi yang mempengaruhi kualitas udara.
2. Perbandingan Polusi Antar Stasiun
 - Stasiun dengan polusi tertinggi: Jika satu stasiun memiliki rata-rata polusi yang lebih tinggi dibandingkan yang lain, maka lokasi tersebut mungkin memiliki sumber pencemaran utama, seperti kawasan industri, lalu lintas padat, atau faktor geografis yang mempengaruhi sirkulasi udara.
 - Stasiun dengan kualitas udara terbaik: Jika ada satu stasiun dengan rata-rata polutan yang lebih rendah secara konsisten, ini bisa mengindikasikan bahwa daerah tersebut memiliki lebih banyak ruang terbuka hijau atau regulasi lingkungan yang lebih ketat.
3. Identifikasi Polutan Dominan
 - Dengan melihat rata-rata polutan untuk setiap stasiun, bisa diketahui polutan mana yang paling dominan di setiap lokasi. Misalnya, jika PM2.5 selalu memiliki konsentrasi tertinggi di semua stasiun, ini berarti polusi udara terutama disebabkan oleh partikel halus, yang sering dikaitkan dengan aktivitas kendaraan dan pembakaran bahan bakar fosil.
 - Jika SO2 atau NO2 tinggi, maka sumber utama polusi kemungkinan berasal dari aktivitas industri atau emisi kendaraan bermotor.
4. Korelasi Antar Polutan
 - Beberapa polutan memiliki korelasi tinggi, misalnya NO2 dan CO seringkali meningkat bersamaan karena keduanya berasal dari emisi kendaraan.
 - O3 (ozon) cenderung meningkat saat NO2 dan CO menurun, karena ozon terbentuk melalui reaksi fotokimia di atmosfer.

## Visualization & Explanatory Analysis

### Bagaimana perbedaan kualitas udara di stasiun "Tiantan" dibandingkan dengan stasiun Shunyi dan Wanliu? dan Apakah lokasi stasiun berpengaruh terhadap dataset yang dihasilkan?:

Saya menggunakan heatmap dari seaborn untuk membandingkan kualitas udara di tiga stasiun. Warna pada heatmap memudahkan saya untuk melihat polutan mana yang lebih dominan di masing-masing stasiun.

```python
print('Visualisasi Perbandingan Kualitas Udara antar Stasiun')
plt.figure(figsize=(12, 6))
sns.heatmap(comparison_df, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Perbandingan Rata-rata Kualitas Udara (2013-2017) antar Stasiun")
plt.xlabel("Polutan")
plt.ylabel("Stasiun")
plt.show()
```

### Bagaimana tren polusi udara di stasiun Tiantan, Shunyi dan Wanliu di 5 tahun terakhir(2013 - 2017)?:

Saya membuat beberapa grafik garis untuk menunjukkan bagaimana tingkat polutan berubah dari tahun ke tahun di setiap stasiun. Dengan cara ini, saya bisa melihat tren apakah polusi semakin meningkat atau justru menurun di setiap lokasi.

```python
print('Visualisasi Tren Polusi Udara 2013-2017')
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Tren Polusi Udara di Tiantan, Shunyi, dan Wanliu (2013-2017)")

for ax, pollutant in zip(axes.flatten(), pollutants):
    ax.plot(st_tiantan_avg.index, st_tiantan_avg[pollutant], label="Tiantan", marker='o')
    ax.plot(st_shunyi_avg.index, st_shunyi_avg[pollutant], label="Shunyi", marker='s')
    ax.plot(st_wanliu_avg.index, st_wanliu_avg[pollutant], label="Wanliu", marker='^')
    ax.set_title(pollutant)
    ax.set_xlabel("Tahun")
    ax.set_ylabel("Konsentrasi (µg/m³ atau mg/m³)")
    ax.legend()
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))
```

**Insight:**
1. Perbandingan Rata-rata Kualitas Udara antar Stasiun (Gambar Pertama)
 - PM2.5 & PM10: Wanliu memiliki konsentrasi rata-rata lebih tinggi dibandingkan dengan Tiantan dan Shunyi, yang bisa berarti daerah ini lebih terpengaruh oleh polusi partikulat.
 - SO2: Wanliu juga memiliki angka SO2 tertinggi, yang bisa menandakan adanya sumber emisi sulfur yang lebih besar, seperti industri atau pembakaran bahan bakar fosil.
 - NO2: Wanliu juga mencatat konsentrasi NO2 tertinggi dibandingkan dua stasiun lainnya, menandakan kemungkinan tingginya polusi dari kendaraan atau aktivitas industri.
 - CO: Semua lokasi memiliki tingkat CO yang tinggi, dengan Wanliu sedikit lebih tinggi dibandingkan dua lainnya.
 - O3: Konsentrasi ozon di Wanliu lebih rendah dibandingkan dengan Tiantan dan Shunyi, yang bisa mengindikasikan perbedaan dalam reaksi fotokimia atau pengaruh polutan lain dalam atmosfer.
 
2. Tren Polusi Udara (Gambar Kedua)
 - PM2.5 & PM10: Tren kedua polutan ini mengalami fluktuasi, dengan penurunan drastis di tahun 2016 dan lonjakan kembali pada tahun 2017. Hal ini mungkin mengindikasikan adanya kebijakan lingkungan atau faktor cuaca yang mempengaruhi kualitas udara.
 - SO2: Konsentrasi SO2 menurun dari 2015 ke 2016, yang bisa menunjukkan upaya pengurangan emisi sulfur, tetapi meningkat lagi di 2017.
 - NO2: Polutan ini menunjukkan tren peningkatan di Wanliu dibandingkan dengan Tiantan dan Shunyi, yang bisa mengindikasikan lebih banyak sumber emisi kendaraan atau industri di wilayah tersebut.
 - CO: Tren karbon monoksida cenderung stabil dari 2013 hingga 2016, tetapi mengalami lonjakan signifikan pada 2017 di semua lokasi.
 - O3 (Ozon): Polutan ini mengalami fluktuasi dengan penurunan tajam di 2017. Ozon sering kali berkorelasi dengan sinar matahari dan reaksi fotokimia, sehingga perubahan ini bisa disebabkan oleh faktor lingkungan atau perubahan emisi precursor ozon.
